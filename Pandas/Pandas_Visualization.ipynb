```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ratings.csv')
df
```

```python
df = df.set_index('Date')
df
```

```python
#using style to set the style of the plot
print(plt.style.available)
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
df.plot() #creating a line plot by default
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title='Ice Cream Ratings')
```

```python
df.plot(kind = 'line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel = 'Scpres')
```

```python
#bar plot
df.plot(kind = 'bar')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
#focusing on a single column
```

```python
df.plot.barh(stacked=True)
#horizontal bar plot
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='yellow')
```

```python
#default histogram
#good to show distribution of variables
df.plot.hist()
```

```python
#specifying number of bins - histogram
df.plot.hist(bins=30)
```

```python
#boxplot
df.boxplot()
```

```python
#default area plot
df.plot.area()
```

```python
df.plot.area(figsize = (10,5))  #specifying figure size
```

```python
#pie chart
df.plot.pie(y='Flavor Rating')
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))  #larger
```
