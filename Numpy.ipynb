```python
#reading a photo as a numpy array
from skimage import io
photo = io.imread('data/nyc.jpg')
type(photo)  #check the type of the photo
```

```python
photo.shape  #check the shape of the photo (height, width, color channels)
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)  #display the photo
```

```python
plt.imshow(photo[0:100, 0:100])  #display a portion of the photo #heig
```

```python
plt.imshow(photo[50:150, 150:200])  #display a different portion of the photo
```

```python
plt.imshow(photo[::-1])  #display the photo upside down
```

```python
plt.imshow(photo[:, ::-1])  #display the photo left to right
#mirror image, reverse the order of the columns
```

```python
plt.imshow(photo[::2, ::2])  #display every second pixel
#every other row and every other column
```

Applying Math functions to Numpy array

```python
#using numpy methods
photo
photo_sin = np.sin(photo)  #taking sine of every element of the photo array
photo_sin
```

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.min(photo))
print(np.max(photo))
print(np.median(photo))
print(np.percentile(photo, 50))  #50th percentile is the median
print(np.argmin(photo))  #index of the minimum value
print(np.argmax(photo))  #index of the maximum value    
```
